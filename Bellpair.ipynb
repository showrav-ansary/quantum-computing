```python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister, BasicAer, execute
from qiskit.visualization import plot_histogram
%matplotlib inline

quantum_register = QuantumRegister(2, 'q') ## Take 2 qubits
classical_register = ClassicalRegister(2,'c') ## Take 2 bits
circuit = QuantumCircuit(quantum_register, classical_register) ## Create a circuit
```

```python
circuit.h(quantum_register[0]) ## Apply Hadamard gate to qubit 0
circuit.cx(quantum_register[0], quantum_register[1]) ## Apply CNOT where qubit 0 is control and qubit 1 is target
circuit.measure(quantum_register[0], classical_register[0]) ## Measure qubit 0 and put in classical register 0
circuit.draw('mpl')
```

```python
backend = BasicAer.get_backend('qasm_simulator') ## Offline virtual simulator
job = execute(circuit, backend, shots=2000) ## Count events in 2000 shots
plot_histogram(job.result().get_counts(), title="|00> vs |01>") ## Plot histogram of results
```
